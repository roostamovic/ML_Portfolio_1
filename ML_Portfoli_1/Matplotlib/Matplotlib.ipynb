```python
# START
# This is Matplotlib Project
# First thing I do is importing libraries
# Let's start
import numpy as np
```

```python
# I also create data using numpy here
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Ok, now we are ready to start
# Let's create a figure first
fig  = plt.figure()

# I also add axes to make it work
ax = fig.add_axes([0,0,1,1])
ax.set_title('Graph')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
# I set all labels
# Now, let's see how my plot looks like
ax.plot(x,y)
```

```python
# Figure in FIGURE
# Let's create a figure that's has another figure in it
fig = plt.figure()

# Here, I should add 2 axes
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

# And let's run our code
ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
```

```python
# Allright, our code has run
```

```python
# It's time to zoom our graph by using matplotlib
# First, I create another figure
fig = plt.figure()

# I add 2 axes again
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

# And name them
ax1.plot(x,z)
ax1.set_title('Solid')
ax1.set_xlabel('Big X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('Small X')
ax2.set_ylabel('Y')

# In order to zoom, I need to use set_xlim and set the interval
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
# Good, you can see zoom of our graph here
```

```python
# I can also save my figure with a format I want, for instance, png, jpg or etc.
fig.savefig('double_plot.png',dpi=111)
```

```python
# I can aslo style my graphs using matplotlib
fig,axes = plt.subplots(nrows=1,ncols=2)
# I can change its color and linestyle
# For example I can change its linestyle into dashed one
# And can change its linewidth
axes[0].plot(x,y,'blue',lw=3,ls='--')
axes[1].plot(x,z,'r',lw=3)
```

```python
# Using figsize I can change area of figure
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))

axes[0].plot(x,y,'blue',lw=4)
axes[1].plot(x,z,'r',lw=3,ls='--')
```

```python
fig.savefig('two_plot.png',dpi=150)
```
