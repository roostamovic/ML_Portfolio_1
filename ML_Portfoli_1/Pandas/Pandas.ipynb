```python
# START
# This project is made by using pandas
# Start with importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import randn
from numpy.random import rand
```

```python
%matplotlib inline
```

```python
# Set style of my graphs
sns.set_style('whitegrid')
plt.style.use('ggplot')
```

```python
# I create a DataFrame using pandas and numpy
df3 = pd.DataFrame(rand(500,4),columns=['a','b','c','d'])
```

```python
# Check the head of it
df3.head()
```

```python
# Now, we can start visualizing our data
# Let's start with creating scatter plot
df3.plot.scatter(x='a',y='b',figsize=(12,3),color='red',s=30)
```

```python
# Second is histogram of our data
df3['a'].plot.hist()
```

```python
# This is a alot more clearer histogram
df3['a'].plot.hist(color='red',alpha=0.5,bins=20)
```

```python
# This is a boxplot of our data
df3[['a','b']].plot.box()
```

```python
# This is styled kdeplot
df3['d'].plot.kde(linewidth=5,linestyle='--')
```

```python
# This is an area plot, it looks like mountain range, doesn't it ?
df3.loc[:30].plot.area(alpha=0.5)
```

```python
# This is aslo area plot but its label is outside of the graph
df3.loc[:30].plot.area(alpha=0.5)
plt.legend(loc='center left',bbox_to_anchor=(1.0,0.5))
```
